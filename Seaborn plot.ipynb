#Distribution Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head(5)
```

```python
sns.set_style('ticks')
sns.set_context('talk')
sns.displot(tips['total_bill'], bins = 40, kde=True,)
plt.xlabel('Total Bill ')
plt.ylabel('Total Count of bill ')
plt.title('Total Bill as per Amount')
plt.tight_layout()
plt.show()
```

```python
sns.set_context('notebook',font_scale =1.05)
sns.set_style('darkgrid')
sns.jointplot(x ='total_bill',y ='tip', data = tips,hue='sex' )
plt.xlabel('Total Bill Amount')
plt.ylabel('Total Tip Amount')
plt.title('Tip as per bill amount')
plt.tight_layout()
plt.show()
```

```python
sns.set_context('talk')
sns.set_style('dark')
sns.pairplot(tips, hue = 'sex', dropna = True, palette = 'coolwarm')
plt.show()
```

#Grid Plot

```python
sns.set_context('talk')
g = sns.PairGrid(tips)
g.map_diag(sns.boxenplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.set_context('notebook')
g = sns.FacetGrid(tips, row='time', col='smoker' )
g.map(sns.histplot, 'total_bill')
```

```python
#sns.set_context('poster')
#sns.despine()
sns.rugplot(tips['total_bill'])
plt.xlabel('Total Bill ')
plt.title('Total Bill as per Amount')
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
```

#Categorial Plot

```python
import numpy as np
sns.set_style('ticks')
sns.barplot(x='sex',y='total_bill', data=tips, estimator = np.std)
plt.xlabel('Gender')
plt.ylabel('Total Bill')
plt.title('Total Bill as per Gender')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='sex', data=tips)
plt.xlabel('Gender')
plt.ylabel('Total Number of People')
plt.title('Total Number of Bill by customer')
plt.tight_layout()
plt.show()
```

```python
#sns.set_style('darkgrid')
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='Reds')
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.ylabel('Total Bill')
plt.title('Total payment as per day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', palette='Greens', split = True)
plt.ylabel('Total Bill')
plt.title('Total payment as per day')
plt.legend(bbox_to_anchor=(1,1), borderaxespad=1)
plt.show()
```

```python
plt.style.use('ggplot')
sns.stripplot(x='day', y='total_bill', data=tips, jitter= True, hue='sex', dodge=True)
plt.legend(bbox_to_anchor=(1,1), borderaxespad=1)
plt.ylabel('Total Bill')
plt.title('Total payment as per day')
plt.show()
```

```python
plt.style.use('seaborn')
sns.set_context('talk',font_scale=1.5)
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.ylabel('Total Bill')
plt.title('Total payment as per day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')    
```

```python
plt.style.use('ggplot')
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
plt.xlabel('Day', fontsize=25)
plt.ylabel('Total Bill', fontsize=25)
plt.show()
```
